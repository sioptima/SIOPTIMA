# PUT/POST API REQUESTS

## 1.) Approve Attendance

// PUT /api/hrd/attendance/att_001/approve
fetch('/api/hrd/attendance/att_001/approve', {
  method: 'PUT',
  headers: {
    'Authorization': 'Bearer eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9...',
    'Content-Type': 'application/json'
  },
  body: JSON.stringify({
    "approvedBy": "hrd_001",
    "notes": "Approved by HRD manager"
  })
})

// RESPONSE:
{
  "success": true,
  "data": {
    "id": "att_001",
    "status": "approved",
    "approvedBy": "hrd_001",
    "approvedAt": "2024-01-15T10:30:00Z",
    "notes": "Approved by HRD manager"
  },
  "message": "Attendance approved successfully"
}
